```python
# Removing some of the most frequent stopwords from the violation_description 
description= df['violation_description'].replace(to_replace = 
                                                 (' IS ',' AT ',' THE ',' OF ',' FOR ','!','#','$',' AND ',' IN ',' OR ',' DO '), 
                                                 value="",regex = True)
```

```python
description
```

```python
text = tuple(description)
```

```python
import collections

def list_to_bigrams(some_list):
    it = iter(some_list)
    old = next(it, None)
    for new in it:
        yield old,new
        old = new
```

```python
a= list_to_bigrams(text);
bigram = list(text)
feature = list_to_bigrams(bigram);
```
